###Plotting in Julia

The package **Gadfly** is used to show how to plot in Julia

```julia
using Gadfly
```

######simulate data

```julia
srand(314)
response  =rand(10)
covariate =rand(10);
```

###<font color='blue'>scatter plot</font>

```julia
plot(x=covariate, y=response, Geom.point)
```

**1. connect points with lines**

```julia
plot(x=covariate, y=response,Geom.point,Geom.line)
```

**2. plot a line with linear fit**

```julia
plot(x=covariate, y=responseGeom.point,Geom.smooth(method=:lm))
```

**3.plot a line with loess**

```julia
plot(x=covariate, y=response,Geom.point,Geom.smooth(method=:loess))
```

**4. plot a horizontal and vertical lines**

```julia
plot(x=covariate, y=response,Geom.point,
yintercept=[0.2, 0.6],Geom.hline,
xintercept=[0.3],Geom.vline(color="red",size=1.5mm))
```

**5. plot a line with intercept 4 and slope 3**

```julia
plot([x->3x+4],0,5)
```

**plot a line with intercept 0.5 and slope 0.25 on the original scatter plot**

```julia
plot(x=covariate, y=response,Geom.point,
layer([x->0.25x+0.5],0,1))
```

###<font color='blue'>Let's make a scatter plot and save it.</font>

```julia
myplot=
plot(x=covariate, y=response,Geom.point,
layer([x->0.25x+0.5],0,1),
Guide.xlabel("Covariate"), 
Guide.ylabel("Response"), 
Guide.title("Plotting"))
```

```julia
draw(SVGJS("myplot.js.svg", 4inch, 3inch), myplot)
```

Now you can check the graph locally.

```julia
plot(x=covariate, y=response, 
  Guide.xlabel("Respondent"), Guide.ylabel("Age"),
Geom.errorbar, ymin=response-1.96*std(response), ymax=response+1.96*std(response),
  Geom.smooth, Geom.point)
```
